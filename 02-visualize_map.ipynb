```python
import folium
```

```python
# 경북대 본관 주소
origin_lat, origin_lng = 35.89049, 128.6120
zoomLevel = 16
```

파일에서 노드 정보와 간선 정보(그림자의 비율)를 읽어서 저장

```python
nodeLatLng = []
with open("nodeInfo.txt", "r") as f :
    for line in f :
        nodeLatLng.append(tuple(map(float, line.split(','))))
```

```python
ratioMatrix = [[0 for _ in range(len(nodeLatLng))] for _ in range(len(nodeLatLng))]
with open("shadowRatio.txt", "r") as f :
    for i, line in enumerate(f) :
        for j, ratio in enumerate(line.split()) :
            ratioMatrix[i][j] = float(ratio)
```

### 노드 정보를 시각화

```python
nodeMap = folium.Map(location=[origin_lat, origin_lng], zoom_start=zoomLevel)

for i, row in enumerate(ratioMatrix) :
    for j, ratio in enumerate(row) :
        if ratio < 0 or i <= j: continue

        folium.PolyLine(
            locations=[nodeLatLng[i], nodeLatLng[j]],
            color='gray',
            weight=5,
            opacity=0.8,
        ).add_to(nodeMap)

for i, (lat, lng) in enumerate(nodeLatLng) :
    folium.CircleMarker([lat, lng], radius=6, fill=True, tooltip=i+1).add_to(nodeMap)

nodeMap.save("01-knuMap.html")
nodeMap
```

### 간선의 그림자 비율을 시각화

```python
def color_weight(ratio) :
    if ratio < 0.1 :
        return 'red'
    elif ratio < 0.5 :
        return 'orange'
    else :
        return 'blue'
    
m = folium.Map(location=[origin_lat, origin_lng], zoom_start=zoomLevel)

for i, row in enumerate(ratioMatrix) :
    for j, ratio in enumerate(row) :
        if ratio < 0 or i >= j: continue

        folium.PolyLine(
            locations=[nodeLatLng[i], nodeLatLng[j]],
            color=color_weight(ratio),
            weight=3
        ).add_to(m)

m
```
